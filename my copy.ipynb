```python
import pandas as pd
file_path =( r'/Users/manyaverma/Downloads/austo_automobile+%282%29+%281%29.csv'
'')
df = pd.read_csv(file_path)
df.head()
```

```python

df.describe(include='all')

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

car_demand_by_age = df.groupby(['Age', 'Make']).size().unstack().fillna(0)

plt.figure(figsize=(10, 6))
sns.heatmap(car_demand_by_age, cmap='coolwarm', annot=True, fmt='.0f')
plt.title('Customer Demand by Age and Car Make')
plt.xlabel('Car Make')
plt.ylabel('Age')
plt.show()


```

```python

df['Salary_bin'] = pd.cut(df['Salary'], bins=[0, 30000, 60000, 100000, 150000], labels=['Low', 'Medium', 'High', 'Very High'])

car_demand_by_salary = df.groupby(['Salary_bin', 'Make']).size().unstack().fillna(0)

car_demand_by_salary.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Car Demand Based on Salary Range')
plt.xlabel('Salary Range')
plt.ylabel('Number of Customers')
plt.show()
```

```python

car_demand_by_dependents = df.groupby(['Marital_status', 'No_of_Dependents', 'Make']).size().unstack().fillna(0)

car_demand_by_dependents.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Car Demand Based on Marital Status and Dependents')
plt.xlabel('Marital Status and No. of Dependents')
plt.ylabel('Number of Customers')
plt.show()
```
